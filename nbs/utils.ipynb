## Duration Formatting

```python
#| export
def format_duration(
    seconds: Optional[float]  # Duration in seconds
) -> str:  # Formatted string (MM:SS or H:MM:SS)
    """Format duration for display in document list and detail views."""
    if seconds is None or seconds < 0:
        return "--:--"
    total_secs = int(seconds)
    hours = total_secs // 3600
    minutes = (total_secs % 3600) // 60
    secs = total_secs % 60
    if hours > 0:
        return f"{hours}:{minutes:02d}:{secs:02d}"
    return f"{minutes}:{secs:02d}"
```

```python
assert format_duration(2537.0) == "42:17"
assert format_duration(3661.0) == "1:01:01"
assert format_duration(65.0) == "1:05"
assert format_duration(0.0) == "0:00"
assert format_duration(None) == "--:--"
assert format_duration(-1.0) == "--:--"
print(f"2537s -> {format_duration(2537.0)}")
print(f"3661s -> {format_duration(3661.0)}")
```

```python
#| export
def format_duration_short(
    seconds: Optional[float]  # Duration in seconds
) -> str:  # Formatted string (e.g., "10.3s")
    """Format duration as compact seconds for average display."""
    if seconds is None:
        return "-.-s"
    return f"{seconds:.1f}s"
```

```python
assert format_duration_short(10.27) == "10.3s"
assert format_duration_short(None) == "-.-s"
print(f"10.27s -> {format_duration_short(10.27)}")
```

## Date Formatting

```python
#| export
def format_date(
    timestamp: Optional[float]  # Unix timestamp
) -> str:  # Formatted date string (e.g., "Feb 19, 2026")
    """Format unix timestamp as a human-readable date."""
    if timestamp is None:
        return "—"
    dt = datetime.fromtimestamp(timestamp, tz=timezone.utc)
    return dt.strftime("%b %d, %Y")
```

```python
assert format_date(1740000000.0) == "Feb 19, 2025"
assert format_date(None) == "—"
print(f"1740000000 -> {format_date(1740000000.0)}")
```

```python
#| export
def format_datetime(
    timestamp: Optional[float]  # Unix timestamp
) -> str:  # Formatted datetime string (e.g., "Feb 19, 2026 12:00")
    """Format unix timestamp as a human-readable date and time."""
    if timestamp is None:
        return "—"
    dt = datetime.fromtimestamp(timestamp, tz=timezone.utc)
    return dt.strftime("%b %d, %Y %H:%M")
```

```python
assert "2025" in format_datetime(1740000000.0)
assert format_datetime(None) == "—"
print(f"1740000000 -> {format_datetime(1740000000.0)}")
```

## Text Formatting

```python
#| export
def truncate_text(
    text: Optional[str],  # Full text to truncate
    max_length: int = 60  # Maximum length before truncation
) -> str:  # Truncated text with ellipsis if needed
    """Truncate text for table and sample display."""
    if text is None:
        return ""
    if len(text) <= max_length:
        return text
    return f"{text[:max_length-3]}..."
```

```python
assert truncate_text("short") == "short"
assert truncate_text("a" * 100, 20) == "a" * 17 + "..."
assert len(truncate_text("a" * 100, 20)) == 20
assert truncate_text(None) == ""
print(f"Truncated: {truncate_text('The art of war is of vital importance to the state. It is a matter of life and death.', 40)}")
```

```python
#| export
def format_time_range(
    start: Optional[float],  # Start time in seconds
    end: Optional[float]     # End time in seconds
) -> str:  # Formatted range (e.g., "0.0s - 2.1s")
    """Format a time range for sample segment display."""
    if start is None or end is None:
        return "-.-s - -.-s"
    return f"{start:.1f}s - {end:.1f}s"
```

```python
assert format_time_range(0.0, 2.1) == "0.0s - 2.1s"
assert format_time_range(None, 2.1) == "-.-s - -.-s"
print(f"Range: {format_time_range(0.0, 2.1)}")
```
